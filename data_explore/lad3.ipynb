```python
import pickle
import numpy as np
import scipy
import gensim
import sklearn
import pandas as pd
```

```python
data = pd.read_csv("../data/sim_train_seg_process.csv", sep="\t", header=None)
data.columns = ["label", "sent1", "sent2"]
data.head(2)
```

```python
doc_list = []

count = 0
for line in (list(data.sent1.values) + list(data.sent2.values)):
    
    doc_list.append(line.split())
    
print("===the len of corpus===", len(doc_list))
print(doc_list[:2])
```

```python
from gensim.models import LdaModel
from gensim import corpora
from gensim.corpora import Dictionary

dictionary = corpora.Dictionary(doc_list)
corpus = [ dictionary.doc2bow(text) for text in doc_list ]
```

```python
for key, value in dictionary.items():
    print(key, value)
    break
print("===dictionary length===", len(dictionary))
print("======", dictionary.get(20))
```

```python
# Set up log to terminal
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

lda = LdaModel(corpus=corpus, id2word=dictionary, num_topics=50, passes=10, eta=0.01)
```

```python
 # print a few most important words for each LDA topic
lda.print_topics(-1) 
```

```python
print("===the topic_size * vocabulari_size===", lda.get_topics().shape)

print("=========get the topic of word================")
# get term topic
print("--word--", dictionary.get(0), "--topic--", lda.get_term_topics(0, minimum_probability=None))

# get document topic
print("=========get the topic of document================")
print("---document content---", " ".join(doc_list[0]))
print("--topic--", lda[corpus[0]])
```

```python
corpus_topics = lda.get_document_topics(corpus)

for document_topic in corpus_topics:
    print(document_topic)
    
    break
```
